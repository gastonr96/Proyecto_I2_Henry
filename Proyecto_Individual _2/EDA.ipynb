```python
def Pasar_Valores(df):
    
    columnas=[ "Provincia", "Tecnologia", "Localidad","Partido"]
    for col in columnas:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: str(x) if pd.notnull(x) else np.nan)
    
    columnas = ["A침o", "Periodo", "Accesos","Trimestre"]
    for col in columnas:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: intentar_convertir_a_int(x))
    
    columnas = ["Velocidad", "Ingresos", "Mbps (Media de bajada)"]  
    for col in columnas:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: intentar_convertir_a_float(x))
    
    return df

def intentar_convertir_a_float(x):
    try:
        return float(x) if pd.notnull(x) else np.nan
    except (ValueError, TypeError):
        return np.nan


def intentar_convertir_a_int(x):
    try:
        return int(x) if pd.notnull(x) else np.nan
    except (ValueError, TypeError):
        return np.nan
    
```

### Trabajamos los outliers para minimizar el error

```python
# Funcionpara eliminar outliers
def Eliminar_outliers(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    # Definimos los l칤mites
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    # Filtramos el DataFrame sin outliers
    df= df[(df[columna] >= limite_inferior) &
    (df[columna] <= limite_superior)]
    return df
```
